```python
## MANE 4962 Homework 3 Question 1
# Load Modules
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import skimage as ski
from sklearn.model_selection import train_test_split
```

```python
# Load and Display Picture
pic = ski.io.imread("seg2.jpg")
plt.imshow(pic)
plt.show()
```

```python
# Properties of Picture
display(pic.shape)
display(pic.size)
# display(pic)
```

```python
# Compiling the Data for KMeans Model to segement
X_data = pic.reshape(-1,3)
X_data = X_data.astype(float) # Changing Data to float
X_data = X_data/255; # Dividing Data by 255
display(X_data.shape)
```

```python
# k-means Model
from sklearn.cluster import KMeans
skmodel = KMeans(
    n_clusters=3,
    init='k-means++',
    n_init='auto',
    max_iter=300,
    tol=0.0001,
    verbose=0,
    random_state=None,
    copy_x=True,
    algorithm='lloyd',
)

skmodel.fit(X_data)
skmodel.labels_ # training labels
skmodel_labels = skmodel.predict(X_data)
skmodel_cluster = skmodel.cluster_centers_
```

```python
# Display Results
print(skmodel_labels)
print(skmodel_labels.shape)
print(skmodel_cluster)
print(skmodel_cluster.shape)
```

```python
# New Picture with cluster center highlighted showing segemented parts
# new = np.uint8(skmodel_cluster)
new_image_data = skmodel_cluster[skmodel_labels.flatten()]
new_pic = new_image_data.reshape(pic.shape)
plt.imshow(new_pic)
plt.show()
```
