```python
"""Load Data Set"""
import tensorflow as tf
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from tensorflow.python.keras.optimizers import *
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# Load Data from Iris Dataset
data = load_iris()

diris = pd.DataFrame()
diris['sepal length'] = data['data'][:,0]
diris['sepal width'] = data['data'][:,1]
diris['petal length'] = data['data'][:,2]
diris['petal width'] = data['data'][:,3]

display(diris) # Verifiy Data 
```

```python
# New Target Features: We can define the new features as Area of the sepal and petal (Length times Width)

### (setosa, versicolor, viriginica) = f(sepal area, petal area)
```

$y = f(X)$

$y \in [0,1,2]$

$X \in {\rm I\!R}^2 $

```python
# Finding Area of the sepal and petal
sepalL = diris.loc[:, 'sepal length']
sepalW = diris.loc[:, "sepal width"]
petalL = diris.loc[:, "petal length"]
petalW = diris.loc[:, "petal width"]

sepalA = sepalL*sepalW
petalA = petalL*petalW

diris = diris.assign(sepal_area = sepalA, petal_area = petalA)
display(diris)

""" Defining X and y """
X = diris.loc[:, ["sepal_area", "petal_area"]]
y = data.target
```

### we are going to split the data and keep some of the data seperate for testing the model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(
    X,y,
    test_size=0.40,
    train_size=0.60,
    random_state=123,
    shuffle=True,
    stratify=y)
```

# import support vector machine model and train it

```python
from sklearn import svm

clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
preds = clf.predict(X_test)
print(preds)
```

```python
print('actual iris species')
print(y_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,clf.predict(X_test) )
print('accuracy score :', acc)
```
