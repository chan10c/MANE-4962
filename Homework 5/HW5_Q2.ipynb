```python
# Load Modules
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import keras
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import axes3d, Axes3D
from matplotlib import cm
import itertools
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from sklearn.model_selection import GridSearchCV 
from scikeras.wrappers import KerasClassifier
from sklearn.preprocessing import LabelEncoder 
from sklearn.preprocessing import StandardScaler 
from tensorflow.keras import datasets, layers, models
from sklearn.datasets import load_breast_cancer # Import Data

# Import Kernels for SVM
from sklearn import svm
```

```python
# Load Data and display related information
health = load_breast_cancer()
# display(data)
display(health.target)
display(health.feature_names)
display(health.target_names)
display(health.data.shape)
```

```python
# Display in Panda Data Frame
df = pd.DataFrame(health.data, columns=health.feature_names)
df["Target"] = health.target
display(df)
```

```python
# The Features that will be inputed to the SVM are  worst compactness, worst concavity, and worst area

display(health.data.shape) # Size 30 correspond to the 30 feature names
X_all, y = health.data, health.target # y-data is already normalized to 0 and 1 

# Accquire the desired features
X = np.column_stack((df['worst compactness'].to_numpy(), df['worst concavity'].to_numpy(), df['worst area'].to_numpy()))  # Pick two datasets
display(X.shape)
```

```python
# Display the desired Features
display(X)
```

```python
# 2a: 
# Build a classification model with a linear kernel
model_linear = svm.SVC(kernel='linear', gamma='auto')
model_linear.fit(X,y)
y_pred = model_linear.predict(X)
linear_acc = accuracy_score(y, y_pred)
print("Test Accuarcy:", linear_acc)
```

```python
# 2b: 
# Build a classification model with a radius basis function
model_radial = svm.SVC(kernel='rbf', C = 2, gamma='auto')
model_radial.fit(X,y)
y_pred2 = model_radial.predict(X)
radial_acc = accuracy_score(y, y_pred2)
print("Test Accuarcy:", radial_acc)
```
